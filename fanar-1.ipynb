```python
def classify_emotion(text):

    prompt = f"""
    You are an expert Arabic emotion classifier.

    Classify the following Arabic sentence into EXACTLY ONE emotion:

    Joy
    Anger
    Sadness
    Fear
    Surprise
    Disgust

    Rules:
    - Return ONLY the emotion label
    - No explanations
    - No extra text

    Sentence:
    "{text}"
    """

    response = client.chat.completions.create(
        model="Fanar-C-2-27B",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        max_tokens=5,
        temperature=0
    )

    result = response.choices[0].message.content.strip()

    return result
```

```python
samples = [
    "أنا سعيد جداً بهذا الخبر",
    "أشعر بحزن شديد اليوم",
    "هذا الشيء مقزز للغاية",
    "أنا خائف من المستقبل",
    "لم أكن أتوقع حدوث ذلك أبداً"
]

for sentence in samples:

    prediction = classify_emotion(sentence)

    print(f"Sentence: {sentence}")
    print(f"Prediction: {prediction}")
    print("-" * 50)
```
